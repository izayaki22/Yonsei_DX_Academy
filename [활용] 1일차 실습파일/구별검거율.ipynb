```python
import numpy as np
import pandas as pd

data = pd.read_csv('02. crime_in_Seoul.csv',encoding='cp949', thousands=',', index_col='관서명')

new_index=[]
for temp in data.index:
  temp = temp[:len(temp)-1]+'구'
  new_index.append(temp)

data.index=new_index

data.loc['종로구']=data.loc['종로구']+data.loc['혜화구']
data.loc['중구']= data.loc['남대문구']+data.loc['중부구']
data.loc['성북구']= data.loc['성북구']+data.loc['종암구']
data.loc['강남구']= data.loc['수서구']+data.loc['강남구']
data.loc['서초구']= data.loc['방배구']+data.loc['서초구']
data.loc['은평구']= data.loc['은평구']+data.loc['서부구']

data = data.drop(['혜화구','남대문구','중부구','종암구','수서구','방배구','서부구'])

data["살인 검거율"]=data["살인 검거"]/data["살인 발생"]*100
data["강도 검거율"]=data["강도 검거"]/data["강도 발생"]*100
data["강간 검거율"]=data["강간 검거"]/data["강간 발생"]*100
data["절도 검거율"]=data["절도 검거"]/data["절도 발생"]*100
data["폭력 검거율"]=data["폭력 검거"]/data["폭력 발생"]*100

data = data[['살인 검거율','강도 검거율','강간 검거율','절도 검거율','폭력 검거율']]

for temp in data.columns:
  data.loc[data[temp]>100, temp] = 100

data.sort_index(axis=0)
```
