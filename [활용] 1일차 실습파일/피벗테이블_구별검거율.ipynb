```python
import numpy as np
import pandas as pd

data = pd.read_csv('02. crime_in_Seoul.csv',encoding='cp949', thousands=',')

data.insert(1,'해당 구',['중구', '종로구','중구', '서대문구', '종로구','용산서','성북구','동대문구', '마포구','영등포구','성동구','동작구', '광진구','서부구', '강북구'
,'금천구', '중랑구', '강남구','관악구','강서구','강동구','성북구','구로구','서초구','양천구'
,'송파구','노원구','서초구','은평구','도봉구','강남구'])

new_data = data.pivot_table(values=['살인 발생','살인 검거','강도 발생','강도 검거','강간 발생','강간 검거','절도 발생','절도 검거','폭력 발생','폭력 검거'],index='해당 구',aggfunc=sum)

new_data["살인 검거율"]=new_data["살인 검거"]/new_data["살인 발생"]*100
new_data["강도 검거율"]=new_data["강도 검거"]/new_data["강도 발생"]*100
new_data["강간 검거율"]=new_data["강간 검거"]/new_data["강간 발생"]*100
new_data["절도 검거율"]=new_data["절도 검거"]/new_data["절도 발생"]*100
new_data["폭력 검거율"]=new_data["폭력 검거"]/new_data["폭력 발생"]*100

new_data = new_data[['살인 검거율','강도 검거율','강간 검거율','절도 검거율','폭력 검거율']]

for temp in new_data.columns:
    new_data.loc[new_data[temp]>100, temp]=100

new_data
```
