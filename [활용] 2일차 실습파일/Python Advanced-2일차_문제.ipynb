#### 001 선 그래프

- 다음의 그림과 같이 plot을 완성하세요.
- 사용된 데이터 범위는 flights 데이터프레임의 passengers 컬럼 시리즈 중 0~11의 인덱스 데이터를 사용했습니다.

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
fig=plt.figure()
ax=plt.subplot(1,1,1)
plt.title('flights')
ax.plot(flights.loc[0:11,'passengers'],
        c='k', ls=':', marker='o', mec='b', mfc='b')
ax.set_ylabel('sepal_length')
ax.set_xticks(range(0,12))
ax.set_xticklabels([i for i in range(1,13)])
plt.show()
````

#### 002 다중 그래프

- 다음의 그림과 같이 plot을 완성하세요.
- 사용된 데이터 범위는 flights 데이터프레임의 passengers 컬럼의 시리즈 중 0 ~ 11의 인덱스 데이터와 12 ~ 23 인덱스 데이터를 사용했습니다.

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
plt.title('flights')
plt.subplot(2, 1, 1)
plt.plot(flights.iloc[0:12, 1], flights.iloc[0:12, 2], c='g',ls='-.', marker='s')
plt.xticks(flights.iloc[0:12,1], labels=['',2,'',4,'',6,'',8,'',10,'',12])
plt.ylabel('sepal_length')
plt.subplot(2, 1, 2)
plt.plot(flights.iloc[12:24, 1], flights.iloc[12:24, 2], c='k',ls='-.', marker='s')
plt.xticks(flights.iloc[0:12,1], labels=['',2,'',4,'',6,'',8,'',10,'',12])
plt.ylabel('sepal_length')
plt.show()
````

#### 003 선 그래프

- 다음의 그림과 같이 plot을 완성하세요.

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
plt.title('iris')
plt.plot(iris.index, iris['sepal_length'],c='b', label='sepal_length')
plt.plot(iris.index, iris['sepal_width'],color='orange',label='sepal_width')
plt.plot(iris.index, iris['petal_length'],c='g', label='petal_length')
plt.plot(iris.index, iris['petal_width'],c='r', label='petal_width')
plt.legend()
plt.show()
````

#### 004 바 그래프

- 다음의 그림과 같이 plot을 완성하세요.

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
import numpy as np

plt.title('iris')
band_width=0.1
index = np.arange(0,5)

plt.bar(index, iris.iloc[0:5,0], band_width, label='sepal_length')
plt.bar(index+band_width, iris.iloc[0:5,1], band_width, label='sepal_width')
plt.bar(index+2*band_width, iris.iloc[0:5,2], band_width, label='petal_length')
plt.bar(index+3*band_width, iris.iloc[0:5,3], band_width, label='petal_width')

plt.xticks(np.arange(band_width*1.5, band_width+5, 1), np.arange(0, 5))
plt.legend()
plt.show()
````

#### 005 파이차트

- titanic 데이터에서 남녀의 수를 구하세요.
- 다음의 그림과 같이 파이차트를 완성하세요. 

-------
<span style="font-weight:normal; color:orange;">Print</span>
image.png

````python
result_dict={}

for temp in titanic['sex']:
    if(temp in result_dict.keys()):
        result_dict[temp]+=1
    else:
        result_dict[temp]=1

plt.title('남녀 성비')
plt.pie(result_dict.values(), labels=result_dict.keys(), autopct='%1.2f%%')
plt.legend()
plt.show()
````
